## One Hot Encoding

Categorical data must be converted to numbers. In our case "Sex" is categorical data.

One hot encoding represents categorical variables as binary vectors. The idea is to convert each category value into a new column and assigns a 1 or 0 (True/False)

```python
values = titanic["Sex"]
values[0:10]
```

```python
# integer encode
titanic = pd.get_dummies(titanic, columns=["Sex"])
```

```python
titanic.head()
```

## Train / Test Split

**Drop the column "Name", because you won't need it.**

```python
titanic = titanic.drop(['Name'], axis=1)
```

**At first separate dependant and independent features!**

(The dependent feature is "Survived")

```python
X = titanic.drop(["Survived"], axis=1)
y = titanic["Survived"]
```

**Create train, and test data**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

# Modeling


We build the decision tree by making use of sklearn.

```python
steps = [
    ('scaler', StandardScaler()),
    ('decision_tree', DecisionTreeClassifier())
]

pipeline = Pipeline(steps)

# train the Decision Tree
clf = pipeline.fit(X_train, y_train)

# prediction for Test data
y_pred = clf.predict(X_test)
```

## Model Evaluation

**Create a confusion matrix**

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

calculate accuracy

```python
accuracy_score(y_true=y_test, y_pred=y_pred)
```

# Visualising a Decision Tree

```python
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO  
from IPython.display import Image  
import pydotplus

dot_data = StringIO()

clf = DecisionTreeClassifier(max_depth=4, random_state=123, max_features=4).fit(X_train, y_train)
export_graphviz(clf, out_file=dot_data,  
                filled=True, rounded=False,
                special_characters=True,class_names=['0','1'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('titanic.png')
Image(graph.create_png())
```
