```python
import pandas as pd
from plotnine import *
```

```python
import os
os.getcwd()
```

```python
filename = "../001_Datasets/Diamonds.csv"
diamonds = pd.read_csv(filename)
```

```python
diamonds.head()
```

# One Variable

## Discrete

```python
(ggplot(data=diamonds) 
  + aes(x = 'cut') 
  + geom_bar()
)
```

## Continuous

```python
(ggplot(diamonds) +
  aes(x = 'price') + 
  geom_density()
)
```

```python
(ggplot(diamonds) +
  aes(x = 'price') + 
  geom_area(stat='bin')
)
```

# Two Variables

## Continuous X, Continous Y

```python
(ggplot(data=diamonds[diamonds['cut'].isin(['Fair', 'Premium'])]) +
  aes(x='x', y='y', color='price', size ='carat') +
  geom_point() +
  facet_grid(['color','cut'])
  #scale_y_log10()
)
```

## Continuous X, Discrete Y

```python
(ggplot(data=diamonds) +
  aes(x='price', y='clarity') + 
  geom_bin2d()
)
```

## Discrete X, Continuous Y

```python
(ggplot(data=diamonds) +
  aes(x='clarity', y='price') + 
  geom_jitter(alpha=.2)
)
```

```python
(ggplot(data=diamonds) +
  aes(x='clarity', y='price') + 
  geom_violin()
)
```

```python
(ggplot(data=diamonds) +
  aes(x='clarity', y='price') + 
  geom_boxplot()
)
```

## Discrete X, Discrete Y

```python
(ggplot(data=diamonds) +
  aes(x='clarity', y='cut') + 
  geom_jitter(alpha=.2)
)
```
